character spletting

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("spech.txt")
docs = loader.load()
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
    separator="\n\n"
)
docs = text_splitter.split_documents(docs)
docs
```

HtmlSplitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string = """
<!DOCTYPE html>
<html>
<head>
    <title>Test Document</title>
</head>
<body>
    <h1>hi</h1>
    <p>This is a paragraph under header 1.</p>
    <h2>gello</h2>
    <p>This is a paragraph under header 2.</p>
</body>
</html>
"""
headers_to_split_on = [
    ("h1" , "Header 1"),
    ("h2" , "Header 2")

]

html_string_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)

html_header_splits  = html_string_splitter.split_text(html_string)
html_header_splits
```

```python
import requests
from langchain_text_splitters import HTMLHeaderTextSplitter

url = "https://en.wikipedia.org/wiki/Artificial_intelligence"

# 1. Get the HTML content
html_content = requests.get(url).text

# 2. Define headers to split on
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4")
]

# 3. Create the splitter
html_string_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)

# 4. Split the HTML content (not the URL!)
html_header_splits = html_string_splitter.split_text(html_content)
html_header_splits
```

```python
import json
import requests
from langchain_text_splitters import RecursiveJsonSplitter

json_data = requests.get("https://jsonplaceholder.typicode.com/posts").json()

json_splitter = RecursiveJsonSplitter(max_chunk_size=300)

# Convert list to dict before splitting
json_chunks = json_splitter.split_json({"root": json_data})

print(json.dumps(json_chunks, indent=2))

```

```python
for chunk in json_chunks[:3]:
    print(chunk)
```

```python
docs = json_splitter.create_documents(json_chunks)
for chunk in docs[:3]:
    print(chunk)
```
