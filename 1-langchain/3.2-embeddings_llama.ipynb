```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = OllamaEmbeddings(
    model = "gemma:2b"
)
```

```python
r1 = embeddings.embed_documents(["Hello world", "Goodbye world"])
```

```python
r1
```

```python
r2 =embeddings.embed_query("Hello world")
r2
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("spech.txt")
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
final_document  = text_splitter.split_documents(docs)

final_document
```

```python
#vector
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_document, embeddings)
db
```

```python
query = "Khalid bin Walid (RA) was appointed governor of Qinnasrin and surrounding regions"
retrived_result = db.similarity_search(query)
print(retrived_result)
```
