```python
#text_loader 
from langchain_community.document_loaders import TextLoader
loader = TextLoader("spech.txt")
loader
```

```python
text_doc = loader.load()
text_doc
```

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("allaboutcats.pdf")
pdf_load  = loader.load()
pdf_load
```

```python
#web based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_paths=["https://lilianweng.github.io/"] 
                       , bs_kwargs=dict(parse_only = bs4.SoupStrainer(class_ = ("post-entry"))
                       ))
web_load = loader.load()
web_load
```

```python
from langchain_community.document_loaders import ArxivLoader

docs = ArxivLoader(query="1706.03762", load_max_docs=2).load()
print(len(docs))
```
